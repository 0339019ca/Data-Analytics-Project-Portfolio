# Clean The Data

### Read the data from the CSV file into a DataFrame

```python
import pandas as pd
```

```python
ramen = pd.read_csv('https://github.com/fenago/dw/raw/main/ramen-ratings.csv')
```

```python
ramen.head()
```

### Display the data to get a count of the number of rows.


```python
ramen.info(verbose=True, memory_usage="deep",show_counts=True)
```

### Rename the Stars column to Rating.


```python
ramen.columns = ramen.columns.str.replace('Stars','Rating')
```

### Convert the Style column to the category type.


```python
ramen.Style = ramen.Style.astype('category')
```

### Drop the country column.


```python
ramen = ramen.drop(columns= ['Country'])
```

### Check that the changes were applied to the DataFrame.


```python
ramen.info()
```

```python
ramen.head()
```

### Rename the Brand column to Company and the Variety column to Product.


```python
ramen_names_dict = {'Brand':'Company','Variety':'Product'}
```

```python
ramen.rename(columns = ramen_names_dict, inplace = True)
```

```python
ramen.head()
```

### Use the duplicated() method to display all rows with duplicate Company/Product combinations.


```python
duplicated = ramen[ramen.duplicated(subset=['Company', 'Product'], keep=False)]
```

```python
duplicated
```

### Drop all rows with duplicate Company/Product combinations except for the first row in each set.

```python
duplicates_drop = ramen.drop_duplicates(keep='first',inplace=True)
```

### Display the data to get a count of the number of rows, and compare that with the original number of rows to see how many were dropped.

```python
duplicates_drop
```

```python
ramen.info()
```

### Drop any rows that have missing values.


```python
ramen_drop = ramen.dropna()
```

### Check again to make sure that all of your changes were applied.


```python
ramen.info()
```
